```python
midpoints = np.concatenate((np.arange(transitmpt, starttime, -period), np.arange(transitmpt, stoptime, period)))
```

```python
def get_KIC(KOI, muirhead_comb):
    return muirhead_comb[muirhead_comb['KOI'] == str(KOI)].KIC.item()
```

```python
def get_CDPPeff(CDPP_N, t_CDPP, t_dur):
    CDPPeff = np.sqrt(t_CDPP/t_dur)*CDPP_N
    print(np.sqrt(t_CDPP/t_dur))
    return CDPPeff
```

```python
def get_SNR_from_CDPP(CDPPeff, tobs, f0, P, depth):
    """Valid for a single transit."""
    
    SNR = np.sqrt((tobs*f0)/P)*(depth/CDPPeff)
    return SNR
```

```python
def get_sigma_individual(SNR, N, Ntransits, tdepth):
    sigma_full = np.sqrt(Ntransits)*(tdepth/SNR)
    sigma_individual = sigma_full*np.sqrt(N)
    return sigma_individual
    
    
def get_N_intransit(tdur, cadence):
    return tduration//cadence
```

```python
# 30 minute cadence
cadence = 0.02142857142857143

tduration = float(ph.get_T14([period], [rprs], [a_rs], [inc]))
```

```python
N = round(get_N_intransit(tduration, cadence))
ntransits = len(midpoints)
depth = 0.036375**2
```

```python

SNRs = np.arange(7, 50, 1)


errbars = get_sigma_individual(SNRs, N, ntransits, depth)
```



## $e = 0.0$, $\omega = 90.0$

```python
def remove_oot_data(time, flux, flux_err, midpoints):
    ttime = []
    tflux = []
    tflux_err = []

    for i in range(len(midpoints)):
        m, b, t1bjd, t1, fnorm, fe1 = ph.do_linfit(time, nflux, flux_err, midpoints[i], 11, 5)
        ttime.append(t1bjd)
        tflux.append(fnorm)
        tflux_err.append(fe1)


    ttime = np.array(ttime).flatten()
    tflux = np.array(tflux).flatten()
    tflux_err = np.array(tflux_err).flatten()

    tflux = np.nan_to_num(tflux, nan=1.0)
    tflux_err = np.nan_to_num(tflux_err, nan=np.nanmedian(tflux_err))
    return ttime, tflux, tflux_err
```

```python

errbar = errbars[0]
SNR = SNRs[0]

directory = 'SNR_' + str(SNR)
if not os.path.exists(directory):
    os.makedirs(directory)

# 30 minute cadence

time = np.arange(starttime, stoptime, cadence)

# Define e and w, calculate flux from transit model
e = 0.0
w = 90.0

flux = ph.integratedlc(time, period, rprs, a_rs, e, inc, w, midpoints[0])

# Adding some gaussian noise based on SNR
noise = np.random.normal(0,errbar,len(time))
nflux = flux+noise

flux_err = np.array([errbar]*len(nflux))

########### Fitting the Transit ###########

# Remove Out of Transit Data

ttime, tflux, tflux_err = ph.remove_oot_data(time, flux, flux_err, midpoints)

### Prior transform and fitting ###

priortransform = [3., 27., 1., 0., 15., 64., 2., 88., 0.1, transitmpt]
nbuffer = 11

# Fitting Transits with Dynesty

dres, pdist, rdist, adist, idist, t0dist = ph.fit_keplc_dynesty(KOI, midpoints, ttime, tflux, tflux_err, priortransform, arrlen, nbuffer, spectplanets, muirhead_comb)


with open('dres.pickle', 'wb') as f:
    pickle.dump(dres, f, pickle.HIGHEST_PROTOCOL)
```

```python
### Saving Dists ###
np.savetxt(str(SNR) + 'pdist.csv', pdist, delimiter=',')
np.savetxt(str(SNR) + 'rdist.csv', rdist, delimiter=',')
np.savetxt(str(SNR) + 'adist.csv', adist, delimiter=',')
np.savetxt(str(SNR) + 'idist.csv', idist, delimiter=',')
np.savetxt(str(SNR) + 't0dist.csv', t0dist, delimiter=',')

# Corner Plot of Transit Fit
fig, axes = dyplot.cornerplot(dres, labels=["period", "Rp/Rs", "a/Rs", "i", "t0"])

plt.savefig(str(SNRs[eindex]) + 'SAtransitcorner.png')

per_f = ph.mode(pdist)
rprs_f = ph.mode(rdist)
a_f = ph.mode(adist)
i_f = ph.mode(idist)
t0_f = ph.mode(t0dist)

# Create a light curve with the fit parameters
fit1 = ph.integratedlc_fitter(ttime, per_f, rprs_f, a_f, i_f, t0_f)

print('Stellar mass (Msun): ', mstar, 'Stellar radius (Rsun): ', rstar)

print('\n')

print('Input params:')
print('Period: ', period)
print('Rp/Rs: ', rprs)
print('a/Rs: ', a_rs)
print('i (deg): ', inc)

print('\n')

print('Fit params:')
print('Period: ', per_f)
print('Rp/Rs: ', rprs_f)
print('a/Rs: ', a_f)
print('i (deg): ', i_f)

T14dist = ph.get_T14(pdist, rdist, adist, idist)
T14errs = ph.get_sigmas(T14dist)

T23dist = ph.get_T23(pdist, rdist, adist, idist)
T23errs = ph.get_sigmas(T23dist)

gs, rho_c = ph.get_g_distribution(rho_star, pdist, rdist, T14dist, T23dist)

g_mean = ph.mode(gs)
g_sigma = np.mean(np.abs(ph.get_sigmas(gs)))

ndim = 2

def loglike(theta):
    """The log-likelihood function."""

    w, e = theta

    model = (1+e*np.sin(w*(np.pi/180.)))/np.sqrt(1-e**2)
    sigma2 = g_sigma ** 2

    return -0.5 * np.sum((g_mean - model) ** 2 / sigma2 + np.log(sigma2))

def unife_prior_transform(utheta):
    """Uniform eccentricity prior"""

    uw, ue = utheta
    w = 360.*uw-90.
    e = 1. * ue

    return w, e


def betae_prior_transform(utheta):
    """Uniform eccentricity prior"""

    uw, ue = utheta
    w = 360.*uw-90.

    a, b = 0.867, 3.03
    e = scipy.stats.beta.ppf(ue, a, b)

    return w, e


### E W fitting with dynesty ###

dsampler = dynesty.DynamicNestedSampler(loglike, unife_prior_transform, ndim=2, bound='multi', sample='rstagger')

dsampler.run_nested()

dres = dsampler.results

truths = [w, e]

fig, axes = dyplot.cornerplot(dres, show_titles=True, title_kwargs={'y': 1.04}, labels=["w", "e"],
                              fig=plt.subplots(2, 2, figsize=(8, 8)))

## Beta-distribution Prior

dsampler = dynesty.DynamicNestedSampler(loglike, betae_prior_transform, ndim=2, bound='multi', sample='rstagger')

dsampler.run_nested()

dres = dsampler.results

np.savetxt(str(SNRs[eindex]) + 'SAdres.csv', dres, delimiter=',')

fig, axes = dyplot.cornerplot(dres, show_titles=True, title_kwargs={'y': 1.04}, labels=["w", "e"],
                              fig=plt.subplots(2, 2, figsize=(8, 8)))

plt.savefig(str(SNRs[eindex]) + 'SAewcorner.png')
```
